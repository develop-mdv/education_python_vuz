#Загрузим новый файл Gold

# Совместный график для UNH (YahooFinance)

```python
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(rows = 3, cols=1)
fig.add_trace(go.Candlestick(x=unh['Date'],
                open=unh['Open'],
                high=unh['High'],
                low=unh['Low'],
                close=unh['Close'],name = 'Свечной график'),row=3, col=1)
fig.add_trace(go.Line(x = unh['Date'], y = unh['Close'],name = 'Цена закрытия'),row=1, col=1)
fig.add_trace(go.Line(x = unh['Date'], y = unh['Open'],name = 'Цена Открытия'),row=1, col=1)
fig.add_trace(go.Bar(x = unh['Date'], y = unh['Open']-unh['Close'],name = 'Барный график'),row=2, col=1)
fig.update_layout(height=1000,
                  width=1000)
fig.update_layout(title="Совместный график для UNH (YahooFinance)")
fig.show()
```

# Совместный график для Gold (Investing.com)

Меняем формат полей для удобного взаимодействия

```python
gold['Макс.'] = gold['Макс.'].apply(lambda x: float(str(x).replace('.','').replace(',','.')))
gold['Мин.'] = gold['Мин.'].apply(lambda x: float(str(x).replace('.','').replace(',','.')))
gold['Откр.'] = gold['Откр.'].apply(lambda x: float(str(x).replace('.','').replace(',','.')))
gold.info()
```

```python
corn['Откр.'] = corn['Откр.'].apply(lambda x: float(str(x).replace('.','').replace(',','.')))
```

```python
fig = make_subplots(rows = 3, cols=1)
fig.add_trace(go.Candlestick(x=gold['Дата'],
                open=gold['Откр.'],
                high=gold['Макс.'],
                low=gold['Мин.'],
                close=gold['Откр.']+gold['Откр.']*gold['Изм. %']/100,
                name = 'Свечной график'),row=3, col=1)
# fig.add_trace(go.Bar(x = gold['Дата'], y = gold['Изм. %'],name = 'Барный график'),row=1, col=1)
# fig.add_trace(go.Bar(x = gold['Дата'], y = gold['Откр.'],name = 'Цена Открытия'),row=1, col=1)
fig.add_trace(go.Bar(x = gold['Дата'], y = gold['Изм. %'],name = 'Барный график'),row=1, col=1)
fig.add_trace(go.Scatter(x = corn['Дата'], y = corn['Откр.'], name='CornArea', stackgroup='one'),row=2, col=1)
fig.add_trace(go.Scatter(x = gold['Дата'], y = gold['Откр.'], name='GoldArea', stackgroup='one'),row=2, col=1)
fig.update_layout(height=1000,
                  width=1000)
fig.update_layout(title="Совместный график для Gold (Investing.com)")
fig.show()
```
